```python
from langchain_community.document_loaders import PDFPlumberLoader
file_path = r"C:\Users\Lenovo\Documents\Project-vs code\Amazon Transcribe\Jupiter-Money\man\517792954-SBI-Card-Statement-0824-22-07-2021.pdf"
loader = PDFPlumberLoader(file_path)
docs = loader.load()
```

```python
import json

# Define a function to parse transaction lines
def parse_transaction(line):
    # Split the line by space delimiter
    parts = line.split()
    # Check if the line has enough parts to extract transaction details
    if len(parts) >= 3:
        # Extract transaction details
        date = parts[0]
        description = ' '.join(parts[1:-1])  # Join middle parts as description
        amount = parts[-1]
        # Return transaction details
        return {
            "Date": date,
            "Description": description,
            "Amount": amount
        }
    else:
        return None

# Extracted text data from PDF
extracted_text = docs[0].page_content

# Define a list to store transaction data
transactions = []

# Iterate over each line of extracted text (excluding the header)
for line in extracted_text.split('\n')[1:]:
    # Parse transaction details
    transaction = parse_transaction(line)
    if transaction:
        # Append valid transaction details to the list
        transactions.append(transaction)

# Convert transaction data to JSON
transactions_json = json.dumps(transactions, indent=4)

# Print the JSON data
print(transactions_json)
```
