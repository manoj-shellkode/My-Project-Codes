```python
from langchain_community.document_loaders import AmazonTextractPDFLoader
import boto3

textract_client = boto3.client("textract", region_name="us-west-2")
textract_features=["TABLES"]
file_path = "s3://my-s3-doc-loader/517792954-SBI-Card-Statement-0824-22-07-2021.pdf"
loader = AmazonTextractPDFLoader(file_path,textract_features,client=textract_client)
docs = loader.load()
```

```python
docs
```

```python
docs[0].page_content
```

```python
import json

extracted_text = docs[0].page_content

transactions = []

for line in extracted_text.split('\n')[1:]:

    parts = line.split('\t')
    if len(parts) >= 3:
        date = parts[0]
        description = parts[1]
        amount = parts[2]

        transactions.append({
            "Date": date,
            "Description": description,
            "Amount": amount
        })

transactions_json = json.dumps(transactions, indent=4)
print(transactions_json)
```
